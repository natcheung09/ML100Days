# 範例重點

如何使用亂數, 資料集來操作


```python
# 載入需要的...
import matplotlib.pyplot as plt
import numpy as np

# 準備數據 ... 假設我要畫一個sin波 從0~180度
# 設定要畫的的x,y數據list....
x = np.arange(0,180)
y = np.cos(x * np.pi / 180.0)

# 開始畫圖
plt.plot(x,y)
# 在這個指令之前，都還在做畫圖的動作 
# 這個指令算是 "秀圖"
plt.show() 

#畫出完整的 sin 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)

plt.plot(x, y_sin)

plt.show()

#畫出完整的 cos 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)

plt.plot(x, y_cos)

plt.show()
```

# 散點圖: Scatter Plots
    
    顏色由（X，Y）的角度給出。

    注意標記的大小，顏色和透明度。

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

T = np.arctan2(Y,X)
plt.scatter(X,Y, s=75, c=T, alpha=.5)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```

```python
#三角函數的arc tangent可以把X Y轉換成0~1的浮點數

#剛好這些浮點數對應到scatter裡面的色碼錶

#才會畫出以上的圖

#你可能要去了解三角函數sin cos tan 還有arcsin arccos arctan才會知道為什麼會算出以上結果

#原來 scatter 色碼可以直接用 float (簡單用 5 個資料點去看)，附個圖也提供給有相同問題的同學。

# scatter參數的資源: https://www.itread01.com/content/1557587069.html
# T是color code
# alhpha是指透明度, 範圍從 0 - 1，數字越大代表越透明。

# https://numpy.org/doc/stable/reference/generated/numpy.arctan2.html
# https://matplotlib.org/3.1.1/api/colors_api.html#module-matplotlib.colors
# https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.scatter.html

n = 5
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
T = np.arctan2(Y,X)
print(T)

plt.scatter(X,Y, s=80, c=T, alpha=.7)
plt.show()


# x y axis area
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
